Reads data 
THRESHOLDS are constants which will be used in the end of the code

```python
import pandas as pd
data = pd.read_csv('/Users/keke/Downloads/PROJECT2 2/cancerTrainingData.csv')
MALIGNANCY_THRESHOLDS = range(5,8)
```

```python
data
```

Malignant data code

```python
malignant_data = data[(data['class'] == 'M')]
malignant_data
```

Avg malignant value code

```python
malignant_means = malignant_data[["radius_mean", "texture_mean", "perimeter_mean","area_mean", "smoothness_mean","compactness_mean", "concavity_mean", "concave_mean", "symmetry_mean","fractal_mean"]].mean()
malignant_means
```

Benign data code

```python
benign_data = data[(data['class'] == 'B')]
benign_data
```

Avg benign value code

```python
benign_means = benign_data[["radius_mean", "texture_mean", "perimeter_mean","area_mean", "smoothness_mean","compactness_mean", "concavity_mean", "concave_mean", "symmetry_mean","fractal_mean"]].mean()
benign_means
```

Midpoints between benign and malignant cells

```python
midpoints = (benign_means+malignant_means)/2
midpoints
```

```python
import pandas as pd
test_data = pd.read_csv('/Users/keke/Downloads/PROJECT2 2/cancerTestingData.csv')
test_data
```

Accuracy finder code

```python
accuracies = []
for threshold in MALIGNANCY_THRESHOLDS:
    correct_count = 0
    false_count = 0
    for index, row in test_data.iterrows():
        comparisons = row[1:-1] > midpoints
        expected_class = "B"
        if comparisons.sum() > threshold:
            expected_class = "M"
        if expected_class == row[-1]:
            correct_count += 1
        else: 
            false_count += 1
    total_count = correct_count+false_count
    accuracy_percent = (correct_count/total_count)*100     
    accuracies.append((threshold,accuracy_percent))
```

```python
best_threshold = max(accuracies, key=lambda x: x[1])
best_threshold
```
